```python
'''Define a class Employee to store information of an employee (empNo, name, department,
basicPay, DA, HRA and grossSalary). Write a constructor which will take as input the
empNo, name, department, basicPay for several employees. The program will calculate the
DA, HRA and total for each employee and define a method to display the details of the
employee having the highest gross salary.'''

class Employee:
    
    emplist=[]
    
    def __init__(self, empNo, name, department,basicpay):
        self.empNo=empNo
        self.name=name
        self.department=department
        self.basicpay=basicpay
        
    def calculate(self, DA, HRA):
        self.grosspay=self.basicpay+self.basicpay*DA/100.0 + self.basicpay*HRA/100.0
        self.DA=self.basicpay*DA/100.0
        self.HRA=self.basicpay*HRA/100.0
        
    def __str__(self):
        return f'Employee name: {self.name}\nEmployee No.: {self.empNo}\nDepartment: {self.department}\nBasic Pay= Rs. {self.basicpay}\nDA: Rs. {self.DA}\nHRA: Rs. {self.HRA}\nGross Pay: Rs. {self.grosspay}'
    
    def getInfo(self):
        print(self)
        __class__.emplist.append(self)
        
    @staticmethod
    def highestSalary():
        __class__.emplist.sort(key=lambda x:x.grosspay, reverse=True)
        print(__class__.emplist[0])
        
if __name__=="__main__":
    
    while True:
        ch=int(input("\nEnter 0 to stop!\nEnter any other number to continue!"))
        if(ch==0):
            print("Process Exited! Thank You!")
            break
        ob=Employee(input("Enter employee number!"), input("Enter employee name!"), input("Enter working department!"), float(input("Enter basic pay!")))
        ob.calculate(float(input("Enter DA percentage!")), float(input("Enter HRA percentage!")))
        ob.getInfo()
        ch1=int(input("Enter 1 to get employee with highest salary!\nEnter any other key to continue!"))
        if(ch1==1):
            print("\nEmployee With highest salary:")
            Employee.highestSalary()
```

```python
'''import re and write scripts to achieve the following:
(i) accept a mobile number as input, and if it starts with 00, replace the 00 with 91
(ii) takes a word as input and satisfy the requirements that it should be of 5 letters,
starts with ‘c’ and ends with ‘r’
(iii) assume a product cataloging system has the following requirements:

first symbol: an uppercase character
second symbol: a digit
third symbol: the special character @
fourth symbol: a lowercase character
write a script that accepts an id and checks whether it is in correct format or not
(iv) check whether a given mail id is in valid format or not.'''

import re

#Part i.
num=input("Enter a mobile number!")
num=re.sub("00", "91", num, 1)
print("Updated mobile number:",num)

#Part ii.
word=input("Enter a word!")
s_word=re.findall("^c.{3}r$", word)
if(len(s_word)>0):
    print("Valid Occurance Found!")
else:
    print("Valid Occurance Not Found!")
    
#Part iii.
id=input("Enter your ID!")
s_id=re.findall("^[A-Z][0-9]@[a-z]$", id)
if(len(s_id)>0):
    print("Valid ID!")
else:
    print("Invalid ID!")
    
#Part iv.
mail=input("Enter your mail ID!")
s_mail=re.findall("^[A-Za-z0-9.]*@[a-z]*mail.com", mail)
if(len(s_mail)>0):
    print("Valid email ID!")
else:
    print("Invalid email ID!")
```
