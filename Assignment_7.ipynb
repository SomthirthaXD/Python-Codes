```python
'''Write a program which can iterate through a list of users (tuples comprising of a name, email
and age) and if the user is at least 18 years old, adds her/him to a dictionary. Write an
exception hierarchy which defines a different exception for each of these error conditions:

• the name is not unique
• the age is not a positive integer
• the user is under 18
• the email format is not valid

Raise these exceptions where appropriate. Whenever an exception occurs, your program
should move onto the next set of data in the list. Print different error messages for each
different kind of exception.'''

class InvalidForDictionaryError(Exception):
    pass

class ListHandler:
    ulist=[]
    def __init__(self, name: str, email: str, age: int):
        self.name=name
        self.email=email
        self.age=age
    def set(self):
        __class__.ulist.append((self.name, self.email, self.age))

import re, traceback as tb
class DictHandler:
    udict={}
    index=1
    @staticmethod
    def set(utup: tuple):
        __class__.udict[__class__.index]=utup
        __class__.index+=1
    @staticmethod
    def check(utup):
        try:
            for k, v in __class__.udict.items():
                if(v[0].__eq__(utup[0])):
                    raise InvalidForDictionaryError("Name not unique!")
            if(utup[2]<0):
                raise InvalidForDictionaryError("Age cannot be negative!")
            elif(utup[2]<18):
                raise InvalidForDictionaryError("Age is below 18!")
            elif(len(re.findall("^[A-Za-z0-9.]*@[a-z]*mail.com$",utup[1]))==0):
                raise InvalidForDictionaryError("Invalid e-mail!")
            else:
                print("User information inserted!")
                DictHandler.set(utup)
        except InvalidForDictionaryError:
            print("Exception caught!")
            tb.print_exc()
    @staticmethod
    def get():
        print("Users' information:")
        for k, v in __class__.udict.items():
            print("User",k,":")
            print("Name:",v[0])
            print("Email:",v[1])
            print("Age:",v[2])
            
if(__name__=="__main__"):
    it=iter(ListHandler.ulist)
    while True:
        ch=int(input("Enter 1 to continue!\nEnter any other number to stop!"))
        if(ch==1):
            ListHandler(input("Enter user name!"), input("Enter user email!"), int(input("Enter user age!"))).set()
            DictHandler.check(next(it))
        else:
            print("Process Exited! Thank You!")
            break
    DictHandler.get()
```

```python
'''Create an OrdinaryCounter class, which mimics a physical clicker counter — the kind people
use at entrances to events and buildings to count how many people are coming in. It may
have an initializer that initializes count to zero, increment, to increment the counter, clear,
to reset the counter to zero, and get_value, to get counter value. Create a SpecialCounter class
that behaves somewhat like the OrdinaryCounter, but always increment by two instead of
one, and also has functionalities to decrement counting.'''

class OrdinaryCounter:
    def __init__(self):
        self.count=0
    def increment(self):
        self.count+=1
    def reset(self):
        self.count=0
    def get_value(self):
        print("Counter value:",self.count)
    def __str__(self):
        return f'Value of count: {self.count}'
    
class SpecialCounter(OrdinaryCounter):
    def __init__(self):
        super().__init__()
    def increment(self):
        self.count+=2
    def decrement(self,val):
        self.count-=val

if(__name__=="__main__"):
    ctype=int(input("Enter 0 for ordinary counter and any other number for special counter!"))
    if(not ctype):
        ob=OrdinaryCounter()
    else:
        ob=SpecialCounter()
    while(True):
        ch=int(input("Enter 1 to increament counter!\nEnter 2 to decrement counter!\nEnter any other number to stop!"))
        if(ch==1):
            ob.increment()
        elif(ch==2):
            if(not ctype):
                print("Ordinary counter cannot be decremented!")
            else:
                ob.decrement(int(input("Enter decrement value!")))
        else:
            print("Counter operations stopped!\n",ob,"\nProcess Exited! Thank You!")
            break      
```

```python
'''Write an “abstract” class, Bag and use it to define some methods which any bag object
should have: add, for adding any number of items to the bag, empty, for taking all the items
out of the bag and returning them as a list, and count, for counting the items which are
currently in the bag. Write a class Item, which has a name attribute and a value attribute –
you can assume that all the items you will use will be Item objects. Now write two subclasses
of Bag which use different underlying collections to store items: ListBag should use a list,
and DictBag should use a dict.
Write a function, repack_bags, which takes any number of bags as parameters, gathers up
all the items they contain, and redistributes them as evenly as possible over all the bags.
Test your code with a ListBag with 20 items, a ListBag with 9 items and a DictBag with 5
items. You should end up with two bags with 11 items each, and one bag with 12 items.'''

from abc import ABC, abstractmethod
class Bag(ABC):
    def __init__(self, collection):
        self.bag=collection
    @abstractmethod
    def add(self, item):
        pass
    def empty(self):
        pass
    def count(self):
        pass

class Item:
    def __init__(self, name, value):
        self.name=name
        self.value=value

import copy
class ListBag(Bag):
    def __init__(self):
        super().__init__([])
    def add(self, item):
        self.bag.append([item.name, item.value])
    def empty(self):
        retlist=copy.deepcopy(self.bag)
        del self.bag[:]
        return retlist
    def count(self):
        print("There are", len(self.bag), "items in the bag!")

class DictBag(Bag):
    def __init__(self):
        super().__init__({})
    def add(self, item):
        self.bag[item.name]=item.value
    def empty(self):
        retlist=[]
        for k,v in self.bag.items():
            retlist.append([k,v])
        self.bag.clear()
        return retlist
    def count(self):
        print("There are", len(self.bag), "items in the bag!")
        
def repack_bags(bags):
    itemlist=[]
    retbags=[]
    for bag in bags:
        if (type(bag)==list):
            for item in bag:
                itemlist.append(item)
        else:
            for k,v in bag.items():
                itemlist.append([k,v])
    totbags=len(bags)
    bagitems=round(len(itemlist)/totbags)
    templist=[]
    c,c1=0,0
    while(c<len(itemlist)):
        templist.append(itemlist[c])
        c+=1
        c1+=1
        if(c1==bagitems):
            retbags.append(copy.deepcopy(templist))
            templist.clear()
            c1=0
    else:
        if(len(templist)>0):
            retbags.append(templist)
    return retbags

def get_item():
    return(Item(input("Enter item name!\n"), input("Enter item value!\n")))

def create_bag():
    ch_collection=int(input("Enter 1 for a list bag!\nEnter any other number for a dictionary bag!\n"))
    if(ch_collection==1):
        bag=ListBag()
    else:
        bag=DictBag()
    while True:
        ch=int(input("Enter 1 to add item to bag!\nEnter 2 to empty bag!\nEnter 3 to count bag elements!\nEnter any other number to stop!\n"))
        if(ch==1):
            bag.add(get_item())
        elif(ch==2):
            print("Items of the bag:",bag.empty(),"\n")
        elif(ch==3):
            bag.count()
        else:
            print("New",type(bag),"bag packed!\n")
            break
    return bag.bag

if(__name__=="__main__"):
    bags=[]
    while True:
        ch=int(input("Enter 1 to pack bag to repack!\nEnter any other number to stop!\n"))
        if(ch!=1):
            print("No more bags will be packed!\n")
            break
        else:
            bags.append(create_bag())
    c=1
    print("The repacked bags are:\n")
    repacked_bags=repack_bags(bags)
    for bag in repacked_bags:
        print("Bag number",str(c)+":",bag)
        c+=1
    print("Process Exited! Thank You!")
```
